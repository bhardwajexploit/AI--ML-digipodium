```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a="https://raw.githubusercontent.com/digipodium/Datasets/main/regression/Salary_Data.csv"
df=pd.read_csv(a)
```

```python
df.sample(5)
df.plot(x='YearsExperience',y="Salary",style="o")
```

```python
df.tail(10)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
reg=LinearRegression()
```

```python
X=df[["YearsExperience"]]
y=df["Salary"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2)
```

```python
reg.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred=reg.predict(X_test)
mse=mean_squared_error(X_test,y_test)
mae=mean_absolute_error(X_test,y_test)
acc=r2_score(y_test,y_pred)
```

```python
print(f"accuracy {acc*100:.1F}")
print(f"mae {mae:.1F}")
print(f"mse {mse:.1F}")
```

- visualize the data


```python
plt.title("linear reg test result")
plt.scatter(X_test,y_test,color='grey',label="actual")
plt.plot(X_test,y_pred,color='red',label="predicted")
plt.legend()
plt.text(10,15,f'Accuracy{acc*100:.1F}',bbox={'edgecolor':'blue'})
plt.show()
```
